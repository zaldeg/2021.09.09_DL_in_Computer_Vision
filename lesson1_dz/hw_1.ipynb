```python
# RGB to HSV
def rgb_to_hsv2(img):
    def hsv_calculate(colors:tuple):
        # receive hsv from rgb for one pixel(3 dims)
        r, g, b = colors
        max_color = max(colors)
        min_color = min(colors)
        if max_color == min_color:
            return (0, 0, max_color)
        if max_color == r and g >= b:
            h = 60 * (g-b) / (max_color-min_color)
        elif max_color == r and g < b:
            h = 60 * (g-b) / (max_color-min_color) + 360
        elif max_color == g:
            h = 60 * (b-r) / (max_color-min_color) + 120
        elif max_color == b:
            h = 60 * (r-g) / (max_color-min_color) + 240
        
        s = 0 if max_color == 0 else 1 - min_color/max_color

        v = max_color

        return [h / 360, s, v]
        
    hsv_img = np.apply_along_axis(hsv_calculate, 2, img.copy())
    
    return hsv_img
```

```python
hsv = rgb_to_hsv2(img)
```
